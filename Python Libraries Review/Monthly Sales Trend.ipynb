Scenario 3: Monthly Sales Trend
You are analyzing sales for an e-commerce company.
Tasks:
Generate daily sales data for 6 months.


Convert it into monthly totals.


Identify:


Best sales month


Worst sales month


Plot:


Monthly sales trend


Highlight best and worst months

```python
# sales data for 6 months
dates = pd.date_range(start="2025-01-01", periods=180)
daily_sales = np.random.randint(500, 5000, len(dates))

df = pd.DataFrame({
    "Date": dates,
    "Sales": daily_sales
})

df.head()
```

```python
# monthly totals
monthly_sales = df.groupby("Month")["Sales"].sum()
print(monthly_sales)
```

```python
# best and worst sales months
best_month = monthly_sales.idxmax()
worst_month = monthly_sales.idxmin()
print("Best Sales Month:", best_month)
print("Worst Sales Month:", worst_month)
```

```python
#Plot monthly sales trend
plt.figure()
monthly_sales.plot()
```

```python
# best and worst months
plt.scatter(best_month.to_timestamp(),monthly_sales[best_month])
plt.scatter(worst_month.to_timestamp(),monthly_sales[worst_month])
plt.xticks(rotation=45)
plt.show()
```
